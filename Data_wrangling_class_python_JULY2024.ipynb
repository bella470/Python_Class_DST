**Data Normalization**

Simple feature saclling

Min-Max

Z-score

```python
#simple feature scalling

mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'] = mydata['width'] / mydata['width'].max()

mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'].describe()
```

```python
#min-max

mydata['length'] = (mydata['length'] - mydata['length'].min()) / (mydata['length'].max() - mydata['length'].min())

mydata[['length', 'width', 'height']].head()
```

```python
mydata['length'].describe()
```

```python
#z-score

mydata['height'] = (mydata['height'] - mydata['height'].mean()) / mydata['height'].std()

mydata[['length', 'width', 'height']].head()
```

```python
mydata['height'].describe()
```

```python
mydata[['length', 'width', 'height']].describe()
```

**Data Normalization (exercise all method accordingly) for norm, curb, horsepower**

```python
mydata_new1[['normalized-losses', 'curb-weight', 'horsepower']].head(10)
```

```python
mydata_new1[['normalized-losses', 'curb-weight', 'horsepower']].describe()
```

```python
#method 1: simple feature scalling

mydata_new1['normalized-losses'] = mydata_new1['normalized-losses'] / mydata_new1['normalized-losses'].max()

mydata_new1[['normalized-losses', 'curb-weight', 'horsepower']].head()
```

```python
mydata_new1['normalized-losses'].describe()
```

```python
#method 2: min -max

mydata_new1['curb-weight'] = (mydata_new1['curb-weight'] - mydata_new1['curb-weight'].min()) / (mydata_new1['curb-weight'].max() - mydata_new1['curb-weight'].min())

mydata_new1[['normalized-losses', 'curb-weight', 'horsepower']].head()
```

```python
mydata_new1['curb-weight'].describe()
```

```python
#method 3: z-score

mydata_new1['horsepower'] = (mydata_new1['horsepower'] - mydata_new1['horsepower'].mean()) / mydata_new1['horsepower'].std()

mydata_new1[['normalized-losses', 'curb-weight', 'horsepower']].head()
```
