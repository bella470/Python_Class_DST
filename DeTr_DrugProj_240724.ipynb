```python
import pandas as pd

df = pd.read_csv("drug200.csv")
df.head(200)
```

```python
X = df.iloc[:, :-1]

X.head()
```

```python
y = df['Drug']
```

```python
X = df.iloc[:, :-1].values
y = df['Drug'].values
```

**Convert Categorical to Numerical**

```python
#Encoding

from sklearn.preprocessing import LabelEncoder

label_sex = LabelEncoder()
X[:, 1] = label_sex.fit_transform(X[:, 1])

label_BP = LabelEncoder()
X[:, 2] = label_BP.fit_transform(X[:, 2])

label_chole = LabelEncoder()
X[:, 3] = label_chole.fit_transform(X[:, 3])

X[:5]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)

X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import jaccard_score

DS = DecisionTreeClassifier()

#fit # training
DS.fit(X_train, y_train)

#predict
y_hat = DS.predict(X_test)

#evaluate
print('The accuracy score is: ', accuracy_score(y_test, y_hat))
print('')
print('The CM is: ', confusion_matrix(y_test, y_hat))
print('')
print('The JS is: ', jaccard_score(y_test, y_hat, average = 'micro')) #average is the class
```
