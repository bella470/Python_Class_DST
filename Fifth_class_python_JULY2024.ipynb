```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

#create numerical data

#normal distribution between that num in the bracket

col1 = np.random.normal(250, 10, 500)
col2 = np.random.normal(100, 20, 500)
col3 = np.random.normal(750, 5, 500)

#create categorical data

#can put any name but to understand just put this first

categorical_data = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

col4 = np.random.choice(categorical_data, 500)
col5 = np.random.choice(categorical_data, 500)

#create date data

start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

delta = end_date - start_date

date_list = [start_date + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 500)

#combine all the columns
#array and in 2D

data = np.column_stack((col1, col2, col3, col4, col5, col6))

#create dataframe

df = pd.DataFrame(data, columns = ['Num1', 'Num2', 'Num3', 'Cat1', 'Cat2', 'Date'])

df.head()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

#create numerical data

#normal distribution between that num in the bracket

#(upper boundry, )

col1 = np.random.normal(10000, 1000, 1000)
col2 = np.random.normal(5000, 100, 1000)
col3 = np.random.normal(500, 10, 1000)
col4 = np.random.normal(1000, 1, 1000)

#create categorical data

#can put any name but to understand just put this first

vid_category = ['Dance', 'Education', 'Comedy', 'Health', 'Sports', 'Technology', 'Cooking', 'Music', 'Lifestyle', 'Art']

col5 = np.random.choice(vid_category, 1000)

#create date data

start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

delta = end_date - start_date

date_list = [start_date + timedelta(days=x) for x in range(delta.days)]

col6 = np.random.choice(date_list, 1000)

#combine all the columns
#array and in 2D

data = np.column_stack((col1, col2, col3, col4, col5, col6))

#create dataframe

df = pd.DataFrame(data, columns = ['Views', 'Like', 'Share', 'Comments', 'Video Category', 'Date'])

df.head()
```

```python
#create bar chart
# Data
categories = ['Dance', 'Education', 'Comedy', 'Lip Sync', 'DIY and Craft']
likes = [3500, 2800, 5000, 1500, 2000]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, likes, color='skyblue')

# Adding titles and labels
plt.title('Likes per TikTok Video Category')
plt.xlabel('Categories')
plt.ylabel('Number of Likes')

# Add the value of each bar on top of the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}', ha='center', va='bottom', fontsize=12)

# Display the plot
plt.xticks(rotation=45)  # Rotate category labels for better readability
plt.tight_layout()
plt.show()
```

```python
#create pie chart

# Data to plot
labels = ['Dance', 'Education', 'Comedy', 'Lip Sync', 'DIY and Craft']
sizes = [25, 35, 20, 10, 10]  # Percentages of each category
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0, 0.1, 0, 0, 0)  # Explode the 1st slice (i.e. 'Dance')

# Plot
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Distribution of TikTok Video Categories')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

# Example data
x = np.linspace(-10, 10, 100)
y = np.sinh(x)

# Symmetrical logarithmic scale on Y-axis
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b')

plt.yscale('symlog')
plt.title('Symmetrical Logarithmic Scale (Y-axis) Plot')

plt.xlabel('X-axis')
plt.ylabel('Y-axis (symlog scale)')
plt.grid(True, which="both")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('fivethirtyeight')

# Make data
x = [1, 1, 2, 2]
y = [1, 2, 1, 2]
z = [0, 0, 0, 0]
dx = np.ones_like(x)*0.5
dy = np.ones_like(x)*0.5
dz = [2, 3, 1, 4]

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.bar3d(x, y, z, dx, dy, dz)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
